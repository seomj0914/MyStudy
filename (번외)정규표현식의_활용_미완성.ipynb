```python
import pandas as pd
```

```python
df = pd.DataFrame(['Mouse','dog','house and parrot','23'],columns=['values'])
df
```

```python
df['v1'] = df['values'].str.contains('dog')
df
```

```python
df['v2'] = df['values'].str.contains('og',regex=False)
df
```

```python
#case=True는 대소문자 구분지어 판별
df['v3'] = df['values'].str.contains('oG',regex=False,case=True)
df
```

```python
#'/d' 는 숫자 
# regex=True 는 즉 '/' 뒤가 regular express 이다.
# regex=False 는 즉 '/' 뒤가 regular express 이 아니다. 일반 string 이다.
df['v4'] = df['values'].str.contains('\d',regex=True)
df['v5'] = df['values'].str.contains('\d',regex=False)
```

```python
df2 = pd.DataFrame(['40','40.0','41','41.0','35'],columns=['values'])
df2
```

```python
df2['v1-1'] = df2['values'].str.contains('.0',regex=True)
df2['v2-2'] = df2['values'].str.contains('\.0',regex=False)
df2['v2'] = df2['values'].str.contains('.0',regex=False)
df2
```
