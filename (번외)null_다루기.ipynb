#null채우기

**주의**
AVG()를 계산할때는 NULL값을 제외하고 계산됨
그러므로 만약 NULL 값이 있을시 일반적인 계산이랑 다른 값으로 출력됨
즉 AVG() 계산 시 NULL값 존재유무 확인해야됨

```python
100,100,100,null,100,100,100

600 / 7 = 100 보다 작음

AVG(~) = AVG (100,100,100,100,100,100) = 100
```

```python
titanic.fillna(0)  # null 대신에 0으로 값 채워넣기
```
