#데이터 이어 붙이기

**SELECT** *

**FROM** air_quality_pm25

**UNION**

**SELECT** *

**FROM** air_quality_no2



```python
#axis=0 위아래로 붙이기,행 붙이기
#axis=1 옆으로 붙이기,열 붙이기
air_quality = pd.concat([air_quality_no2,air_quality_pm25],axis=0)
air_quality
```

```python
#데이터 잘 이어붙어졌는지 확인 
air_quality['parameter'].value_counts()
```

```python
#데이터 잘 이어붙어졌는지 확인 
air_quality['parameter'].unique()
```

```python
print(air_quality_no2.shape)
print(air_quality_pm25.shape)
print(air_quality.shape)
```

```python
#keys= 는 index하나를 추가해서 데이터의 출처 적어줌
air_quality = pd.concat([air_quality_no2,air_quality_pm25],axis=0,keys=["NO2","PM25"])
```

```python
# PM25 값 데이터만 추출하기
air_quality.loc['PM25']
```

```python
air_quality
```

```python
#index에 있는 값을 column으로 만들어주고 싶을때 -> level_0 ,level_1 생성 
#모든 인덱스를 컬럼으로 변형
air_quality.reset_index()
```

```python
#'level_0' index 만 column 으로 만들어주고 싶을때
#지정 인덱스만 컬럼으로 변형
air_quality.reset_index(level = 0 )
```

#조인(JOIN)

조인하고자 하는 각 테이블의 기준컬럼 이름이 같을때



```python
stations_coord = pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/air_quality_stations.csv")
stations_coord.head(2)
```

```python
stations_coord['location'].drop_duplicates()
```

```python
air_quality.head()
```

```python
air_quality['location'].drop_duplicates()
```

```python
#left = air_quality, right = station_coord 붙이기
#left join 
#location 이라는 컬럼을 기준으로
air_quality= pd.merge(air_quality, station_coord, how='left', on='location')
air_quality.head()
```
