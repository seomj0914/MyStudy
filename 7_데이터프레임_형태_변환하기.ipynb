#Long to wide table format
##pivot
- index : index 색인으로 사용될 컬럼

- columns : column 색인으로 사용될 컬럼

- values : value에 채우고자 하는 컬럼

```python
#column 추가하기
no2 = air_quality[air_quality['parameter'] == 'no2']
```

```python
no2_subset = no2.sort_index().groupby(['location']).head(2)
no2_subset
```

```python
#column은 'location' , index는 그대로 , 값은 'value' 로 변횐
no2_subset.pivot(columns='location',values='value')
```

```python
no2.pivot(columns='location',values='value')
```

```python
no2.pivot(columns='location',values='value').plot(figsize = (12,6))
```

#Pivot table
- index : index 색인으로 사용될 컬럼

- columns : column 색인으로 사용될 컬럼

- aggfunc : 데이터 축약 시 사용할 함수 (mean, sum, count ...)

- margins : 행/열별 총 합

```python
air_quality.pivot_table(index='location',columns='parameter',values='value',aggfunc='mean',margins=True)
```

```python
#위의 코드랑 똑같은 연산 간편하게 
air_quality.groupby(['location','parameter'])['value'].mean()
```

#Wide to long format

-

```python
#reset_index() 는 index에 있는 값들 column으로 변경 
no2.pivoted = no2.pivot(columns='location',values='value').reset_index()
no2.pivoted
```

```python
#위의 코드와 반대되는 연산
# 'date.utc' 제외한 나머지 columns 들 BETR801,	FR04014	,London Westminster 행으로 재배치
no2.pivoted.melt(id_vars='date.utc')
```

```python
#지정한 column만 행으로 재배치
#'date.utc' 제외한 나머지 columns 들 BETR801,	FR04014	,London Westminster 중에서 BETR801데이터만 뽑아서 행으로 재배치 
no2.pivoted.melt(id_vars='date.utc', value_vars=['BETR80'])
```
