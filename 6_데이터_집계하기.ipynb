#그룹 별 집계
-SELECT Sex , AVG(Age)

-FROM titanic

-GROUP BY Sex

```python
#1개의 컬럼을 기준으로 GROUP BY
titanic.groupby('Sex')['Age'].mean()
```

```python
# #여러개의 컬럼을 기준으로 GROUP BY
## 'Pclass','Sex' 별로 'Age' 평균 구하기
titanic.groupby(['Pclass','Sex'])['Age'].mean()
```

#그룹 별 데이터 규모 파악하기

```python
titanic.groupby('Pclass')['PassengerId'].count()
```

```python
titanic['Pclass'].value_counts()
```

```python
titanic.info()
```

```python
#null값 제거해서 계산
#1번 Pclass는 AGE 데이터가 있는 사람들은 186명
#2번 Pclass는 AGE 데이터가 있는 사람들은 173명
titanic.groupby('Pclass')['Age'].count()
```

```python
# 단순히 행의 갯수 계산 ,null값 포함 
titanic.groupby('Pclass')['Age'].size()
```

```python
#value_counts는 null값 제외하고 계산하는게 디폴트
#null값 포함해서 계산하고 싶을때 dropna=False
titanic['Age'].value_counts(dropna=False).sum()
```

```python
##추가 여러개 셀 보고싶을때
titanic.groupby('Age').agg({"Pclass":['min','mean']
                           ,"PassengerId":['min','mean']})
```
