```python
import pandas as pd
```

```python
titanic = pd.read_csv('https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/titanic.csv')
titanic.head(2)
```

#소문자 변형

```python
#titanic['Name'].str.upper()
titanic['Name'].str.lower()
```

#문자열 자르기

```python
titanic['Name Splited'] = titanic['Name'].str.split(',')
titanic.head(5)
```

```python
#각 리스트의 0번째 값 가져오기
titanic['Surname'] = titanic['Name Splited'].str.get(0)
titanic.head(2)
```

```python
type(titanic['Name'].str.split(','))
```

```python
#위와 동일한 결과
titanic['Name'].str.split(',').str.get(0)
```

```python
# Countess 라는 글자가 들어간 열은 True
titanic['Name'].str.contains('Countess')
```

```python
# True 에 해당되는 것만 도출
titanic[titanic['Name'].str.contains('Countess')]
```

#문자열 길이 계산하기

```python
titanic['Name Length'] = titanic['Name'].str.len()
```

```python
titanic
```

```python
#이름 길이값이 가장 긴 것은 True
titanic['Name Length'] == titanic['Name Length'].max()
```

```python
#이름 길이값이 가장 긴 값 도출
titanic[titanic['Name Length'] == titanic['Name Length'].max()]
```

```python
#'Name Length'가 가장 긴 값의 인덱스 값(번호)
titanic['Name Length'].idxmax()
```

```python
titanic.loc[titanic['Name Length'].idxmax()]
```

```python
#titanic['Name Length'].idxmax() 의 'Name'값만 도출
titanic.loc[titanic['Name Length'].idxmax(),'Name']
```

#문자열 바꾸기

```python
#series 는 여러개를 한번에 바꿀 수 있음.
titanic['Sex'].replace({'male':'M','female':'F'})
```

```python
#한번에 바꿀 수 없음
#순서 중요함(밑에 예시 참조)
titanic['Sex Short v1'] = titanic['Sex'].str.replace('female','F')
titanic['Sex Short v1'] = titanic['Sex Short v1'].str.replace('male','M')
titanic[['Sex','Sex Short v1']]
```

```python
#순서 중요함
#female도 male이라는 단어가 포함되어있어 'feM'로 입력값 도출

titanic['Sex Short v2'] = titanic['Sex'].str.replace('male','M')
#titanic['Sex Short v1'] = titanic['Sex Short v1'].str.replace('male','M')
titanic[['Sex','Sex Short v2']]
```
